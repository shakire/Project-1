```python
import mysql.connector
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="tonystark",
    auth_plugin='mysql_native_password')
mycursor = mydb.cursor()
mycursor.execute("CREATE DATABASE Orders")
```

```python
mycursor.execute("USE Orders")

  
create_table_query = """
    CREATE TABLE Order1 (
        order_id INT PRIMARY KEY,
        order_date DATE,
        ship_mode VARCHAR(255),
        segment VARCHAR(255),
        country VARCHAR(255),
        city VARCHAR(255),
        state VARCHAR(255),
        postal_code VARCHAR(255),
        region VARCHAR(255)
    )
    """
mycursor.execute(create_table_query)

file_path = "C:/Users/USER/null_filled1.csv"
data = pd.read_csv(file_path)   

insert_query = """
INSERT INTO Order1 (order_id, order_date, ship_mode, segment, country, city, state, postal_code, region)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

for index, row in data.iterrows():
    mycursor.execute(insert_query, (
        row['order_id'],
        row['order_date'],
        row['ship_mode'],
        row['segment'],
        row['country'],
        row['city'],
        row['state'],
        row['postal_code'],
        row['region']
    ))

mydb.commit()
```

```python
mycursor.execute("USE Orders")
mycursor.execute("DROP TABLE IF EXISTS order2")
create_table = """create table order2 (
                    order_id INT,
                    category VARCHAR(100),
                    sub_category VARCHAR(100),
                    product_id VARCHAR(100),
                    cost_price INT,
                    list_price INT,
                    quantity INT,
                    discount_percent INT,
                    discount_price INT,
                    selling_price INT,
                    profit INT )"""

mycursor.execute(create_table)

file_path = "C:/Users/USER/null_filled2.csv"
data = pd.read_csv(file_path) 

insert_value = """INSERT INTO order2 (order_id,category, sub_category, product_id, cost_price, list_price, quantity, discount_percent, discount_price, selling_price, profit)
                  VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""

for index, row in data.iterrows():
    mycursor.execute(insert_value, (
        row['order_id'],
        row['category'],
        row['sub_category'],
        row['product_id'],
        row['cost_price'],
        row['list_price'],
        row['quantity'],
        row['discount_percent'],  
        row['discount_price'],
        row['selling_price'],
        row['profit']
    ))

mydb.commit()



        

                                        
```
